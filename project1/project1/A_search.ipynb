```python
import numpy as np
import random
import time

def attacked_queens_pairs(seqs):

    a = np.array([0] * 81)
    a = a.reshape(9, 9)
    n = 0

    for i in range(1, 9):
        if seqs[i-1] != 0:
            a[seqs[i - 1]][i] = 1
    for i in range(1, 9):
        if seqs[i - 1] == 0:
            continue
        for k in list(range(1, i)) + list(range(i + 1, 9)):
            if a[seqs[i - 1]][k] == 1:
                n += 1
        t1 = t2 = seqs[i - 1]
        for j in range(i - 1, 0, -1):
            if t1 != 1:
                t1 -= 1
                if a[t1][j] == 1:
                    n += 1
            if t2 != 8:
                t2 += 1
                if a[t2][j] == 1:
                    n += 1
        t1 = t2 = seqs[i - 1]
        for j in range(i + 1, 9):
            if t1 != 1:
                t1 -= 1
                if a[t1][j] == 1:
                    n += 1
            if t2 != 8:
                t2 += 1
                if a[t2][j] == 1:
                    n += 1
    return int(n/2)

def display_board(seqs):

    board = np.array([0] * 81)
    board = board.reshape(9, 9)

    for i in range(1, 9):
        board[seqs[i - 1]][i] = 1
    print('chessboard:')
    for i in board[1:]:
        for j in i[1:]:
            print(j, ' ', end="")
        print()
    print('attack pairs: ' + str(attacked_queens_pairs(seqs)))
```

```python
start = time.time()
queue = [{'queens':8, 'pairs':28, 'seqs':[0] * 8}]
ans = []
flag = 0

while queue:
    first = queue.pop(0)
    if first['pairs'] == 0 and first['queens'] == 0:
        ans = first['seqs']
        flag = 1
        break
    nums = list(range(1, 9))
    seqs = first['seqs']
    if seqs.count(0) == 0:
        continue
    for j in range(8):
        possition = seqs.index(0)
        seqtemp = list(seqs)
        temp = random.choice(nums)
        seqtemp[possition] = temp
        nums.remove(temp)
        queue.append({'queens':seqtemp.count(0), 'pairs':attacked_queens_pairs(seqtemp),'seqs':seqtemp})
    queue = sorted(queue, key=lambda x:(x['pairs']+x['queens']))

if ans:
    print('find the sequence：' + str(ans))
    display_board(ans)
else:
    print('not find the solution')

end = time.time()
print('time:' + str('%.2f' % (end-start)) + 's')
```
