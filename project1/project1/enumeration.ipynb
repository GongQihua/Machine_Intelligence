```python
import json

seq = [[i, j, k, l, m, n, o, p]
       for i in range(1, 9)
       for j in range(1, 9)
       for k in range(1, 9)
       for l in range(1, 9)
       for m in range(1, 9)
       for n in range(1, 9)
       for o in range(1, 9)
       for p in range(1, 9) 
       if all([i!=j, i!=k, i!=l, i!=m, i!=n, i!=o, i!=p,
               j!=k, j!=l, j!=m, j!=n, j!=o, j!=p,
               k!=l, k!=m, k!=n, k!=o, k!=p,
               l!=m, l!=n, l!=o, l!=p,
               m!=n, m!=o, m!=p,
               n!=o, n!=p,
               o!=p])]
print('total ' + str(len(seq)) + ' possible sequence')
with open('seq.json', 'w') as file_seq:
    json.dump(seq, file_seq)
```

```python
import numpy as np
import time

start = time.time()
with open('seq.json', 'r') as file_seq:
    seq = json.load(file_seq)

solutions = 0

for s in seq:
    a = np.array([0] * 81)
    a = a.reshape(9,9)
    
    flag = 1
    
    for i in range(1, 9):
        a[s[i-1]][i] = 1
    
    for i in range(1, 9):
        t1 = t2 = s[i-1]
        for j in range(i-1, 0, -1):
            if t1 != 1:
                t1 -= 1
                if a[t1][j] == 1:
                    flag = 0
                    break
            
            if t2 != 8:
                t2+= 1
                if a[t2][j] == 1:
                    flag = 0
                    break
        
        if flag == 0:
            break
        
        t1 = t2 = s[i-1]
        for j in range(i+1, 9):
            if t1 != 1:
                t1 -= 1
                if a[t1][j] == 1:
                    flag = 0
                    break
            
            if t2 != 8:
                t2+= 1
                if a[t2][j] == 1:
                    flag = 0
                    break
        
    if flag == 1:
        solutions += 1
        print('the ' + str(solutions) + 'th solution，sequence is：'+ str(s) + ' chessboard：')

        for i in a[1:]:
            for j in i[1:]:
                print(j, ' ',end="")
            print()

        print('---------------')
print('total ' + str(solutions) + ' solution')
end = time.time()
print('time:' + str('%.2f' % (end-start)) + 's')
```
