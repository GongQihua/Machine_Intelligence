```python
import json

seq = [[i, j, k, l, m, n, o, p]
       for i in range(1, 9)
       for j in range(1, 9)
       for k in range(1, 9)
       for l in range(1, 9)
       for m in range(1, 9)
       for n in range(1, 9)
       for o in range(1, 9)
       for p in range(1, 9)
       if all([i != j, i != k, i != l, i != m, i != n, i != o, i != p,
               j != k, j != l, j != m, j != n, j != o, j != p,
               k != l, k != m, k != n, k != o, k != p,
               l != m, l != n, l != o, l != p,
               m != n, m != o, m != p,
               n != o, n != p,
               o != p])]

print('total ' + str(len(seq)) + ' possible sequence')
with open('seq.json', 'w') as file_seq:
    json.dump(seq, file_seq)
```

```python
import numpy as np

def attacked_queens_pairs(seqs):

    a = np.array([0] * 81)
    a = a.reshape(9, 9)
    n = 0

    for i in range(1, 9):
        a[seqs[i - 1]][i] = 1

    for i in range(1, 9):
        for k in list(range(1, i)) + list(range(i + 1, 9)):
            if a[seqs[i - 1]][k] == 1:
                n += 1
        t1 = t2 = seqs[i - 1]
        for j in range(i - 1, 0, -1):
            if t1 != 1:
                t1 -= 1
                if a[t1][j] == 1:
                    n += 1
            if t2 != 8:
                t2 += 1
                if a[t2][j] == 1:
                    n += 1
        t1 = t2 = seqs[i - 1]
        for j in range(i + 1, 9):
            if t1 != 1:
                t1 -= 1
                if a[t1][j] == 1:
                    n += 1
            if t2 != 8:
                t2 += 1
                if a[t2][j] == 1:
                    n += 1
    return int(n/2)

def display_board(seqs):

    board = np.array([0] * 81)
    board = board.reshape(9, 9)

    for i in range(1, 9):
        board[seqs[i - 1]][i] = 1
    print('chessboard:')
    for i in board[1:]:
        for j in i[1:]:
            print(j, ' ', end="")
        print()
```

```python
import random
import time

start = time.time()
with open('seq.json', 'r') as file_seq:
    seqs = json.load(file_seq)

cur = random.choice(seqs)

while True:
    ans = []
    count = 0
    dicts = []
    a = list(range(1,9))
    for item in cur:
        for i in a[:item - 1] + a[item:]:
            seqs_tmp = list(cur)
            seqs_tmp[count] = i
            ans.append(seqs_tmp)
        count = count + 1
        if count == 8:
            break
    for s in ans:
        tmp_pair = attacked_queens_pairs(s)
        dicts.append({'seqs':s, 'attacked_queens_pairs':tmp_pair})
    nums = []
    for d in dicts:
        nums.append(d['attacked_queens_pairs'])
    mins = min(nums)
    current_attacked_queens_pairs = attacked_queens_pairs(cur)
    if mins >= current_attacked_queens_pairs:
        answer = cur
        break
    temp = []
    for d in dicts:
        if d['attacked_queens_pairs'] == mins:
            temp.append(d['seqs'])
    cur = random.choice(temp)

if attacked_queens_pairs(answer) == 0:
    print('find the global optimal：' + str(answer))
    display_board(answer)
else:
    print('not find global optimal, local optimal：' + str(answer))
    print('attack pairs:'+ str(attacked_queens_pairs(answer)))
    display_board(answer)
end = time.time()
print('time:' + str('%.2f' % (end-start)) + 's')
```
